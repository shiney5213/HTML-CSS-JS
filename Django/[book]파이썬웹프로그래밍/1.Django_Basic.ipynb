# 1. 장고 개발의 기본 사항 및 가상환경 사용하기

#### 장고의 장점
- MVT(Model Veiw Template)에 따라 진행.단축 함수, 제네릭 뷰 -> 쉽고 빠름
- 서드파티에 의해 외부 라이브러리가ㅏ 풍부

## 1.1. 장고 개발의 기본 사항
### 1.1.1. MVT 개발방식
> 모델
- 테이블 정의
- models.py

> 뷰
- 에플리케이션의 제어 흐름 및 처리 로직 정의
- views.py

> 템플릿
- 사용자가 보가 될 화면의 모습을 정의
- *.html

#### 장점
- 모델, 뷰, 템블릿간의 독립성 유지
- 느슨한 결합(Loose Coupling)
- 디자이너, 응용 개발자, db설계자 간 협업 용이

### 1.1.2 MVT 코딩 순서
- 화면 설계는 뷰와 템플릿 코딩으로 연결
- 테이블 설게는 모델 코딩에 반영
- 모델을 먼저 코딩하고 뷰와 템플릿 코딩이 일반적

> 일반적인 순서
- 프로젝트 뼈대 만들기: 프로젝트 및 앱 개발에 필요한 디렉터리와 파일 생성
- 모델 코딩하기: 테이블 관련 사항개발(models.py, admin.py)
- URLconf 코딩: URL 및 뷰 매핑 관계 정의(urls.py)
- 뷰 코딩: 애플리케이션 로직 개발(view.py)
- 템플릿 코딩: 화면 UI 개발(templates/html)

### 1.1.3. settings.py 주요 사항
- 프로젝트 설정 파일

> 필수 수정
- 데이터베이스 설정: SQLite3 지정
- 애플리케이션 등록: 개발하는 앱 모두 등록
- 템플릭 항목 설정
- 정정파일항목 설정: STATIC_URL
- 타임존 지정(최초: 세계표준시 -> 한국 시간으로 변경)


### 1.1.4. models.py 주요 사항
- 테이블을 정의하는 파일

#### ORM(Object Relational Mapping)
- 테이블을 클래슬 매핑해서 테이블에 대한 CRUD(Create,Read, Update,Delete)기능을 클래스 객체에 대해 수행하면 장고가 내부적으로 SQL 처리를 하여 데이터베이스에 반영해주는 방식
- 테이블을 하나의 클래스로 정의
- 테이블 컬럼은 클래스의 함수로 매핑
- django.db.models.model 클래스 상속받아 정의

#### 마이그레이션(migrations)
- 테이블 및 필드의 생성, 삭제, 변경 등 db에 대한 변경사항을 알려주는 정보

### 1.1.5 URLconf 주요 사항
- urls.py
> 프로젝트 전체 URL을 정의하는 프로젝트 URL + 앱마다 정의하는 앱 URL
- 2계층으로 나누어 코딩하는 방법
- 변경이 쉽고 확장이 용이

> <font color = 'red'>URL 패턴별 이름 공간</font>
- reverse()함수나 {% url %}탬플릿 태그를 사용해 소스에 URL을 하드코딩하지 않아도 필요한 URL을 추출하는 기능

### 1.1.6 views.py 주요 사항
- 가독성, 유지보수 편리성, 재활용 고려해서 로직을 코딩해야 함.

#### 종류: 둘다 사용
> 함수형 뷰

> 클래스형 뷰
- 장고가 제공하는 제네렉 뷰를 사용할 수 있음.
- 재활용, 확장성 굿


### 1.1.7 templates 주요 사항
- 웹페이지별로 템플릿 파일 필요
- 템플릿 디렉터리에 모아서 보관

#### 종류
> 프로젝트 템플릭 디렉터리
- TEMPLATES 설정에 DIRS항목에 지정된 디렉터리
- base.html등 전체 프로젝트의 룩앤필(look and feel)관련 파일 모아놓기

> 앱 템플릿 디렉터리
- 애플리케이션 디렉터리마다 존제하는 templates/디렉터리

### 1.1.8. Admin 사이트
- 테이블의 내용을 열람, 수정하는 기능 제공
- 테이블에 있는 콘텐츠를 편집


### 1.1.9. 개발용 웹서버 (run server)
- 개발과정에서 작성된 코드를 실행하고 테스트하기 위한 테스트용 웹 서버

#### <font color = red>Apache/ Nginx</font>
- 실제로 고객에게 제공하기 위한 상용 웹 서버
