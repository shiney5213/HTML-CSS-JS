# 7. 프로젝트2:라이트박스
> 라이트박스
- 팝업형태로 나타나는 모달 윈도(modal window)의 한 종류
- 사진 또는 이미지 감상에 사용
- 이미지를 감상하는 UI(대화 상자)가 팝업으로 들 때 배경을 어두워지게 처리해 대화 상자에 집중할 수 있음.

> 라이트박스 UI
- 팝업창은 화면에 고정
- 가운데 집중해야 한다는 시각적인 알림 제공
- 라이트박스가 닫히ㅈ기 전가지 배경 부분 컨트롤 불가

## 7.1. 개요
> 작업 순서
- 1. HTML과 CSS

- 3. 배경화면 처리 및 인터랜션 금지
- 4. 인디케이터 스타일링
