# 1. SQL 이해하기

## 1.1. 데이터 베이스 기본
> 데이터베이스 
- 정리된 방식으로 데이터를 저장하는 공간

> DBMS(DataBase Management System)
- 데이터베이스 소프트웨어

> 테이블
- 구조화된 특정한 타입의 데이터 목록
- 특정한 타입의 데이터를 저장할 수 있는 구조화된 파일
- 테이블에 저장되는 데이터는 한가지 타입이거나 목록
- 모든 테이블을 다른 테이블과 구별되는 고유한 이름을 가짐.
- 테이블명이 고유한 이유: 데이터베이스명, 테이블명등 여러가지 항목이 조합되어 만들어지기 때문
    
> 스키마: 데이터베이스와 테이블 구조, 속성에 대하한 정보
- 테이블: 어떤 데이터가 저장되는지를 정의하는 특성과 속성을 가짐
- 특성과 속성: 어떤 데이터가 저장되는지, 어떻게 나눠지는지, 각 정보에는 어떤 이름을 갖는지등의 정보를 포함
- 전체 데이터베이스 뿐 아니라 데이터베이스에 있는 특정 테이블을 표현하는데 사용
- 테이블 간의 관계를 표현

> 컬럼
- 테이블에 있는 하나의 필드, 모든 테이블은 한개 이상의 컬럼으로 구성
- 각 컬럼은 특정한 정보를 담음(주소, 고객번호, 이름 등)

> 데이터 나누기
- 여러개의 컬럼에 데이터를 올바르게 나누어 넣는 것
- 요구사항에 따라 데이터의 특성을 정하여 나누기

> 데이터 타입
- 컬럼에 저장할 수 있는 데이터의 유형
- 숫자형, 문자형 등
- 허용되는 데이터의 유형
- 모든 컬럼은 데이터타입을 가지고 있음
- 컬럼에는 정해진 데이터타입만 허용됨

> 데이터 호환성
- 테이터 타입, 데이터 타입명은  SQL이 호환되지 않게 하는 주요 원인 중 하나

> 행
- 각 레코드는 각가의 행에 담기게 됨
- 테이블의 가로
- 테이블에 있는 레코드

> 주 키(Primary key)
- 테이블의 각 행을 구별하는 컬럼
- 주키는 특정 행을 참조하기 위해 사용하는데, 주 키 없이 테이블에 있는 특정 행을 업데이트하거나 삭제하기 힘듦.
- 항상 주 키를 정의해야 함(주키를 생성하여 미래의 데이터를 조작하고 관리할 수 있도록 보장해야 함.)
- 반드시 1개의 컬럼만 정의할 필요는 없음. 

> <font color = 'red'>주키의 조건</font>
- 같은 주 키 값을 갖는 행이 1개만 존재한다
- 모든 행은 주 키값을 가져야 한다(NULL값을 허용하지 않음.)
- 주 키 컬럼에 있는 값은 수정되거나 업데이트될 수 없다
- 주 키 값은 재사용될 수 없다(삭제된 경우에는 재사용 가능)

## 1.2. SQL이란 무엇인가?
- 구조화된 질의 언어(Structured Query Language)
- Sequel(시퀄)로 발음.
- 데이터베이스와 소통하기 위한 언어
- 몇 개의 단어로만 구성-> 간단하면서도 효율정

> 장점
- 대부분의 DBMS가 지원
- 매우 쉬움
- 문법은 단순하지만, 다양한 구문을 사용해서 복잡하면서도 정교한 데이터베이스 작업 수행 가능

> SQL확장
- 표준 SQL: ANSI SQL
- PL SQL
- Transact-SQL
