# 3. Data processing using arrays

> 배열을 사용한 데이터 처리
- Numpy 배열을 사용하면 반복문을 작성하지않고 간결한 배열여산을 통해 많은 종류의 데이터 처리 작업 가능
- 벡터화: 배열 연산을 사용해서 반복문을 명시적으로 제거하는 기법
- 일반적으로 벡터화된 배열에 대한 산술연산은 순수 파이선에 비해 2~3배, 수백배까지 빠름

```python
# 그리드에서 두 배결을 받아 (x, y)짝을 만들기

# 1000 equally spaced points
points = np.arange(-5, 5, 0.01)
points
```

```python
#행단위와 열단위로 각각 해당 배열을 정방(square) 행렬로 선언
xs, ys = np.meshgrid(points, points)
```

```python
xs
```

```python
ys
```

```python

from matplotlib.pyplot import imshow, title
import matplotlib.pyplot as plt
```

```python
z = np.sqrt(xs ** 2 + ys **2)
z
```

```python
plt.imshow(z, cmap = plt.cm.gray)
plt.colorbar
```

```python
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
```

 ## Expressing conditional logic as array operations
 
 > 배열 연산으로 조건절 표현하기

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
```

```python
# zip: index 기준으로 배열을들 묶어서 사용할 수 있음.
result = [(x if c else y) for x, y, c in zip(xarr, yarr, cond)]
result
```

```python
# np.where(조건, 참값, 거짓값)
result = np.where(cond, xarr, yarr)
result
```

```python
arr = randn(4,4)
arr
```

> np.wherea()
- where에 넘긴 배열을 같은 크기의 배열이거나 스칼라값일 수 있다.
- where를 사용하면 좀 더 복잡한 연산을 수행할 수 있다.

```python
np.where(arr>0, 2, -2)
```

```python
# set only positive values to 2
np.where(arr > 0, 2, arr)
```

## Methods for boolean arrays

> 불리언 배열을 위한 메소드
- 불리언 값은 1(True), 0(False)으로 취급
- 불리언 배열에 따른 sum 메소드를 실행하면 True의 원소의 개수를 반환
- any() : 하나 이상의 True값이 있는지 검사
- all(): 모든 원소가 True인지 검사

```python
arr = randn(100)
arr
```

```python
(arr>0).sum()
```

```python
bools = np.array([False, False, True, False])
bools
```

```python
# 하나라도 True인가?
bools.any()
```

```python
# 모두 True인가?
bools.all()
```
