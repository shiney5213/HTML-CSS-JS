## 1. 넘파이 사용

```python
#sin 함수

import math
import matplotlib.pyplot as plt

T = range(100)
X = [(2 * math.pi * t) / len(T) for t in T]
Y = [math.sin(value) for value in X]

plt.plot(X, Y)

plt.show()
```

```python
# 이항식 그리기
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-3, 2, 200)
Y = X**2 - X + 1

plt.plot(X, Y)
plt.show()
```

## 다중 곡선 그리기


```python
# [0, 2*pi] 간격의 sin, cos 둘다 그리기

X = np.linspace(0, 2 * np.pi, 100)
Ya = np.sin(X)
Yb = np.cos(X)

plt.plot(X, Ya)
plt.plot(X, Yb)
plt.show()
```

```python
# 지연 랜더링, deferred rendering 메커니즘은 맷플랏립의 핵싱


def plot_slope(X, Y):     # 그래프의 기울기를 부르는 함수
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:], Ys/Xs)
    
X = np.linspace(-3, 3, 100)
Y = np.exp(-X**2)   #  1/ e **x**

plt.plot(X, Y)
plot_slope(X, Y)
plt.show()
```

## 파일 데이터로부터 곡선 그리기

```python
# 컬럼이 2개일 때

# my_data.txt
'''
0  0
1  1
2  4
4 16
5 25
6 36
'''
```

```python
# 파일 읽어오기
f = open('my_data.txt', 'w')
f.write('0  0')
f.write('\n1  1')
f.write('\n2  4')
f.write('\n4 16')
f.write('\n5 25')
f.write('\n6 36')
f.close()


f = open('my_data.txt')
text = f.read()
print(text)
f.close
```

```python
X, Y  = [],[]
for line in open('my_data.txt', 'r'): # 파일에서 한 줄씩 읽어오기
    values = [float(s) for s in line.split()]  #  공백을 기준으로 나누기
    X.append(values[0])
    Y.append(values[1])
    
plt.plot(X, Y)
plt.show()
```

```python
with open('my_data.txt','r') as f:
    X, Y = zip(*[[float(s) for s in line.split()] for line in f])
    # zip 이용해서 X, Y값을 한꺼번에 묶끼
    # * : 묶여져 있는 요소 풀기
    
plt.plot(X, Y)
plt.show()
```

## 막대 차트 그리기

```python
# pyplot.bar(): 막대 차트 전용 함수
data = [5,25,50,20]

plt.bar(range(len(data)), data)  # 데이터 개수만큼 x축 개수(indxe) 지정
plt.show()
```

```python
# 막대 굵기: width
data = [5,25,50,20]

plt.bar(range(len(data)), data, width = 1)  
plt.show()
```

```python
# 수평 막대 : plt.barh

data = [5,25,50,20]

plt.barh(range(len(data)), data)  # 데이터 개수만큼 x축 개수(indxe) 지정
plt.show()
```

```python
# 다중 막대 차트 그리기

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

X = np.arange(4)

plt.bar(X + 0.00, data[0], color = 'b', width = 0.25)
plt.bar(X + 0.25, data[1], color= 'g', width = 0.25)
plt.bar(X + 0.50, data[2], color = 'r', width = 0.25)

plt.show()
```

```python
# 반복자 enumerate는 현재 행과 해당 첨자를 둘다 반환

data = [[5., 25., 50., 20.],
        [4., 23., 51., 17.],
        [6., 22., 52., 19.]]

color_list = ['b','g','r']
gap = .8/len(data)

for i, row in enumerate(data): # 인덱스과 값을 모두 가져옴
    X = np.arange(len(row))
    print(X)
    plt.bar(X + i * gap, row,
           width = gap,
           color = color_list[i % len(color_list)])
    
plt.show()
 
```

# 분할 막대 차트 만들기

```python
# bottom 인자 사용

import matplotlib.pyplot as plt

A = [5., 30., 45., 22.]
B = [5., 25., 50., 20.]

X = range(4)

plt.bar(X, A, color = 'b')
plt.bar(X, B, color = 'r', bottom = A)
plt.show()
```

```python
# 값을 두 집합 이상으로 분할
A = np.array([5., 30., 45., 22.])
B = np.array([5., 25., 50., 20.])
C = np.array([1.,  2.,  1.,  1.])

X = range(len(A))

plt.bar(X, A, color = 'b')
plt.bar(X, B, color = 'g', bottom = A)
plt.bar(X, C, color = 'r', bottom = A + B)
plt.show()
```

```python
# 값을 두 집합 이상으로 분할
data = np.array([[5., 30., 45., 22.],
                 [5., 25., 50., 20.],
                 [1.,  2.,  1.,  1.]        
    ])

color_list = ['b','g','r']

X = np.arange(data.shape[1])  # X : array 형태!

for i in range(data.shape[0]):
#     print(data[i])                
#     print(np.sum(data[:i], axis = 0))
#     print(color_list[i % len(color_list)])
          
    plt.bar(X, data[i],
           bottom = np.sum(data[:i], axis = 0),       #  아래에 놓일 데이터 선택
           color = color_list[i % len(color_list)])    # 나머지로 인덱스 결정 -> 색 가져오기
    
plt.show()
```

## 삼각화 그리기


```python
# 공간 위치를 처리할 때 삼각화 생긴다. 점과 이웃 관계 사이의 거리를 보여줌
import matplotlib.tri as tri

data = np.random.rand(100,2)

# print(data[:,0])
# print(data[:,1])
triangles = tri.Triangulation(data[:,0], data[:,1])

plt.triplot(triangles)
plt.show()
```

```python
# 삼각화 plot은 scatter plot을 선으로 연결한 것!!
plt.scatter(data[:,0], data[:,1])
plt.show()
```
