```python

# color=('b','g','r','c','m','y','k','w')    # colors : A sequence of matplotlib color args through which the pie chart
y
```

```python
plt.figure(figsize = (6,6))
plt.pie(y)
plt.show()
```

```python
# label값 지정하기
label = ['Blue', 'Orange', 'Green','Red','Violet', 
         'Brown','Pink', 'Gray','Yello', 'Cyan']
plt.figure(figsize=(6,6))
plt.pie(y, labels=label)    # labels : A sequence of strings providing the labels for each wedge
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
# 인자값(ratio,explode,  labels, shadow, startangle, autopct)
# ratio: 각 범주가 데이터에서 차지하는 비율
# explode: 파일 조각이 돌출되는 크기-> 각 범주별로 돌출되는 비율 지정
# labels: 범주 전달
# shadow: 그림자 여부
# startangle: 시작 각도(default: 0도)
# autopct: 파이 조각의 전체 대비 백분율 => 자동으로 계산해중!!!


labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes  = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# 빵의 주요 성분
# 그래프, 범례 영역을 나누기 위해 subplots 사용
fig, ax = plt.subplots(figsize=(10, 12), subplot_kw=dict(aspect="equal"))

recipe = ["375 g 밀가루",
          "75 g  설탕",
          "250 g 버터",
          "300 g 딸기"]

data = [float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda x :func(x, data),  
                                  textprops=dict(color="w"))
# plt.axes.Axes.pie(data, autopct, textprops)
# data : data
# autopct: 숫자의 레이블
# textprops: 텍스트에 전달할 인수=> 색 지정


ax.legend(wedges, ingredients,
          title="주요성분",
          loc="center left", 
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=14, weight="bold")
#plt.setp(obj, size, weight) -> set propobility


ax.set_title("빵의 주요 성분")

plt.show()
```

## 6. Histogram 작성

```python
# 정규 분포를 갖는 그래프
data = np.random.normal(5, 3, 1000) # (평균, 표준편자, 개수)
```

```python
plt.figure(figsize = (10,4))
plt.plot(data)
plt.show()
```

```python
# 위에 그린 y축을 기준으로 구간별로 잘라서 x축에 표현
plt.figure(figsize=(10,4))
plt.hist(data)
plt.show()
```

```python
# bins값 지정-> 빈수가 많을수록 정규분포곡선에 가까워짐
plt.figure(figsize=(10,4))
plt.hist(data, bins=20)     # bins : integer, If an integer is given, `bins + 1` bin edges are returned
plt.show()
```

```python
# alpha: 투명도, histtype
plt.figure(figsize=(10,4))
plt.hist(data,bins=10, facecolor='red', 
         alpha=0.4, histtype='step') # 단게별로 채워지게
plt.hist(data,bins=20, facecolor='green', 
         alpha=0.4, histtype='stepfilled')     # histtype : {'bar', 'barstacked', 'step',  'stepfilled'}
plt.show()

#histtype: 히스토그램의 유형
# - bar: 막대가 나란히 배려
# - barstacked: 여러 데이터가 쌓여있는 형태
# - step: 채워지지 않은 라인 plot
# - stepfilled: 채워진 plot

# 아래쪽 그래프의 두구간을 합친 것이 뒤의 그래프의 한 구간의 값과 같음.-> 그래프의 높낮이 차이가 남
```

```python
x = np.random.randn(1000)
plt.figure(figsize = (10, 4))
plt.plot(x)
plt.show()
```

```python
# 정규분포를 따르는 hist rmflrl
# normed = 1: 히스토그램의 면적의 합 = 1
plt.figure(figsize=(10,4))
plt.hist(x, normed=1, bins=100)  
plt.show()
```

## 7. box 그래프 작성


```python
#np.random.normal() : 정규분포를 따르는 데이터에서 뽑기
# loc : 평균
# scale : 표준편차
# size : 개수
s1 = np.random.normal(loc=0,  scale=1.0, size=1000)    # loc : Mean ("centre") of the distribution.
s2 = np.random.normal(loc=5,  scale=0.5, size=1000)    # scale : Standard deviation (spread or "width") of the distribution.
s3 = np.random.normal(loc=10, scale=2.0, size=1000)    # size : Output shape
```

```python
plt.figure(figsize=(10,4))
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```
