# 서울시 구별 와이파이존 빈도수 시각화

> 서울시 구별 무료와이파이존의 수를 지도에 나타낸다.


```python
import pandas as pd
import json
import warnings
warnings.simplefilter(action = 'ignore', category = FutureWarning) # warning 창 무신
```

```python
# 구와 구의 좌표 정보가 있는 json 파일 -> 지도에 표시할 수 있음.
geo_path = './data/skorea_municipalities_geo_simple.json'
geo_str  = json.load(open(geo_path, encoding='utf-8'))
```

```python
# 구, 경로, 와이파이 개수를 가져와야 함.
df = pd.read_csv('data/seoul_wifi_zone.csv')
df.head()
```

```python
df.columns = (['구', '유형', '위치명', '위도', '경도', '설치회사'])
df.head()
```

```python
# 각 구별로 몇 개인지 확인
freq_gu = df['구'].value_counts()
freq_gu
```

```python
freq_gu.describe()
```

```python
map = folium.Map(location = [37.5102, 126.982], zoom_start = 11)
map
```

```python
map = folium.Map(location =[37.5102, 126.982], tiles = 'Stamen Toner', zoom_start=11)
map
```

```python
map = folium.Map(location =[37.5102, 126.982], tiles = 'Stamen Toner', zoom_start = 11)

map.choropleth(geo_data = geo_str,
              data = freq_gu,
              columns = ['gu', 'counts'],
              fill_color= 'YlGnBu',  #PuRd, YlGnBu
              key_on = 'feature.id')
map
```
